```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a,b,n = 5,7,10
```

```python
x = np.array([i for i in range(1,n+1)])
y = a*x//b - a*(x//b)

plt.figure(figsize=(3,4))
plt.plot(x,y, label='y')


x = np.array([i for i in range(1,n+1)])
y = a*x//b
plt.plot(x,y)
    
    
    
x = np.array([i for i in range(1,n+1)])
y =  - a*(x//b)
plt.plot(x,y)

plt.legend()
plt.show()
```
